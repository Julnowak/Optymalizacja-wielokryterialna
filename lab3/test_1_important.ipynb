### Import bibliotek

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

### Definicja funkcji celu

```python
def objectives(u):
    x, y = u
    f1 = x**2 + y**2
    f2 = (x - 1)**2 + (y + 1)**2
    return np.array([f1, f2])


U = [(-1, 1), (-1, 1)]  # Dziedzina U - kwadrat

# Generacja zakresu dla funkcji
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)

# Funkcje nieliniowe
f1 = x**2 + y**2
f2 = (x - 1)**2 + (y + 1)**2
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data for the surface
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
f1 = x**2 + y**2  # The function f1 = x^2 + y^2

# Create a figure and 3D subplot
fig = plt.figure(figsize=(10, 5))
ax1 = fig.add_subplot(121, projection='3d')

# Plot the surface
surf = ax1.plot_surface(x, y, f1, cmap='viridis', edgecolor='k', alpha=0.8)

# Add title and labels
ax1.set_title('f1 = x^2 + y^2')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('f1(x, y)')

# Adding a color bar to indicate the values
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=10)

# Display the plot
plt.tight_layout()
plt.show()
```

### Skalaryzacja przez Funkcj Liniow

```python

# Skalaryzacja przez funkcj liniow
def scalarization_linear(u0, lambd):
    def scalarized_objective(u):
        return np.dot(lambd, objectives(u))
    result = minimize(scalarized_objective, u0, bounds=U)
    
    return result

# Generuj zakres wartoci lambda
lambda_values = np.linspace(0.05, 0.95, 30)
solutions = []

for lambda1 in lambda_values:
    lambd = np.array([lambda1, 1 - lambda1])
    res = scalarization_linear(u0=[0, 0], lambd=lambd)
    solutions.append((res.x, objectives(res.x)))

# Wyodrbnij wartoci funkcji celu
F_values = np.array([sol[1] for sol in solutions])

# Wykres
# Generuj przestrze funkcji celu

X, Y = np.meshgrid(x, y)
U_grid = np.column_stack([X.ravel(), Y.ravel()])
F_all = np.array([objectives(u) for u in U_grid])

plt.figure(figsize=(10, 7))
plt.scatter(F_all[:, 0], F_all[:, 1], s=1, label='Dostpne F(u)')
plt.scatter(F_values[:, 0], F_values[:, 1], color='red', label='Punkty Pareto (Skalaryzacja)')
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('Skalaryzacja przez funkcj liniow')
plt.legend()
plt.grid(True)
plt.show()
```

### Metoda -ogranicze

```python
# Optymalizacja metod \(\epsilon\)-ogranicze
def epsilon_constraint_optimization(epsilon):
    def objective(u):
        return objectives(u)[0]  # Minimalizuj F1(u)
    def constraint(u):
        return epsilon - objectives(u)[1]  # F2(u) <= epsilon
    constraints = {'type': 'ineq', 'fun': constraint}
    result = minimize(objective, x0=[0, 0], bounds=U, constraints=constraints)
    return result.x, objectives(result.x)

# Generuj zakres wartoci epsilon
epsilon_values = np.linspace(min(F_all[:, 1]), max(F_all[:, 1]), 75)
solutions = []

for epsilon in epsilon_values:
    res_u, res_F = epsilon_constraint_optimization(epsilon)
    solutions.append((res_u, res_F))

# Wyodrbnij wartoci funkcji celu
F_values = np.array([sol[1] for sol in solutions])

# Wykres
plt.figure(figsize=(10, 7))
plt.scatter(F_all[:, 0], F_all[:, 1], s=1, label='Dostpne F(u)')
plt.scatter(F_values[:, 0], F_values[:, 1], color='red', label='Punkty Pareto (Metoda \(\epsilon\)-ogranicze)')
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('Metoda \(\epsilon\)-ogranicze')
plt.legend()
plt.grid(True)
plt.show()
```

### Skalaryzacja przez Odlego od Wybranego Punktu Dominujcego

```python
# Skalaryzacja przez odlego od punktu dominujcego z wagami komponentowymi
def scalarization_distance(u0, x_dominating, lambd, p_norm):
    def distance_objective(u):
        diff = objectives(u) - x_dominating
        weighted_diff = lambd * np.abs(diff)
        return np.linalg.norm(weighted_diff, ord=p_norm)
    
    result = minimize(distance_objective, u0, bounds=U)
    return result

# Generujemy wszystkie dostpne wartoci funkcji celu w dziedzinie U
x_vals = np.linspace(-1, 1, 100)
y_vals = np.linspace(-1, 1, 100)
X, Y = np.meshgrid(x_vals, y_vals)
U_grid = np.column_stack([X.ravel(), Y.ravel()])
F_all = np.array([objectives(u) for u in U_grid])

# Wybieramy punkt dominujcy (powinien dominowa wszystkie F(u))
dominating_point = np.array([5.0, 5.0])  # Upewnij si, e dominuje
p_norm = 2  # Norma euklidesowa

# Generujemy rne wektory wag lambda
lambda_values = np.linspace(0.05, 0.95, 30)  # Unikamy wartoci 0 i 1
solutions = []

for lambda1 in lambda_values:
    lambda_vector = np.array([lambda1, 1 - lambda1])
    res = scalarization_distance(u0=[0, 0], x_dominating=dominating_point, lambd=lambda_vector, p_norm=p_norm)
    solutions.append((res.x, objectives(res.x), lambda_vector))

# Wyodrbniamy wartoci funkcji celu dla punktw Pareto
F_values = np.array([sol[1] for sol in solutions])

# Wykres
plt.figure(figsize=(10, 7))
plt.scatter(F_all[:, 0], F_all[:, 1], s=1, label='Dostpne F(u)')
plt.scatter(F_values[:, 0], F_values[:, 1], color='red', label='Punkty Pareto (Skalaryzacja przez odlego)')
plt.xlabel('F1')
plt.ylabel('F2')
plt.title('Skalaryzacja przez odlego od punktu dominujcego')
plt.legend()
plt.grid(True)
plt.show()
```
